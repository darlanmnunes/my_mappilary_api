```python
from mapillary_api import *
```

```python
# read the sample region
sample_region = gpd.read_file('sample_region.geojson')

sample_bbox = sample_region.geometry[0].bounds
```

```python
# using the basic function to fetch mappillary data:
sample_mappillary_md = get_mapillary_images_metadata(*sample_bbox,outpath='sample_metadata.json')

sample_as_gdf = mapillary_data_to_gdf(sample_mappillary_md,outpath='sample_metadata.geojson')
```

```python
# download all the images in the sample:
sample_outfolderpath = 'sample_images'

create_dir_if_not_exists(sample_outfolderpath)

download_all_pictures_from_gdf(sample_as_gdf,sample_outfolderpath)
```

```python
# downloading in the vicinity of a Point:
buffer = 5

p = Point(-49.275853153000583, -25.444585293776242)
buffer = radius_to_degrees(buffer, p.y)
p_bounds = p.buffer(buffer).bounds
# using the basic function to fetch mappillary data:
p_metadata = get_mapillary_images_metadata(*p_bounds)
p_md_gdf = mapillary_data_to_gdf(p_metadata)
download_all_pictures_from_gdf(p_md_gdf,sample_outfolderpath)
```
